```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import path, sep
%matplotlib inline
```

```python
# load data
dataPath = path.join("data", "Consumer_Complaints.csv")
compDat = pd.read_csv(dataPath)
```

```python
# chop small part to test
# testSet = compDat[:100]
testSet = compDat
```

```python
# convert date from str to dt
dateAsDate = pd.to_datetime(testSet['Date received'])
```

```python
# groupby and plot
# .unstack(fill_value=0) is from https://stackoverflow.com/a/45495293/4010864
plotSet = testSet.groupby([dateAsDate.dt.year, 'Product']).size().unstack(fill_value=0)
plotSet.plot.bar(figsize=(18,10))
```
